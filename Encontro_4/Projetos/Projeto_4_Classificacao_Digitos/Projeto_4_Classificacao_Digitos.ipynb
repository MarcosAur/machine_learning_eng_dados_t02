# Projeto 4: Classificação de Dígitos por Imagem (Multiclasse)

## Objetivo
Classificar imagens em miniatura (8x8 pixels achatadas em 64 colunas de intensidade) de dígitos escritos à mão em sua classe correta (dígitos de 0 a 9). A meta de negócio é obter uma **Acurácia Balanceada superior a 0.90 (Balanced Accuracy > 0.90)** no conjunto de teste.

## Por que Acurácia Balanceada?
Em problemas multiclasse, se uma classe é muito comum e outras muito raras, a Acurácia padrão pode mascarar uma péssima performance em classes raras. A **Balanced Accuracy** calcula a acurácia (Recall) de cada classe individualmente e tira a média simples, garantindo que o modelo seja robusto para prever todos os dígitos de 0 a 9 de forma homogênea.

## Desafios
1. **Engenharia de Pipelines para Imagens:** A entrada é uma matriz achatada de 64 pixels. Precisamos aplicar escala uniforme de forma isolada via pipeline.
2. **Análise de Matriz de Confusão Multiclasse:** Analisar uma matriz 10x10 para identificar quais dígitos o modelo está confundindo mais (ex: confundir 3 com 8, ou 1 com 7).
3. **Deploy:** Serializar o modelo e escrever um simulador de inferência de produção que aceite um array/vetor plano de 64 pixels correspondente a um novo dígito desenhado.
